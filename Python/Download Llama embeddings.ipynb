```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
from datasets import load_dataset, Dataset
```

```python
from datasets import load_dataset, Dataset
ds = load_dataset("domingue-lab/question_difficulty_embedded-full", split="train", use_auth_token=True)
```

```python
df = pd.json_normalize(ds)
```

```python
embeddings = np.array(df['embedding'].tolist())
```

```python
embeddings.shape
```

```python
embedding_df = pd.DataFrame(df['embedding'].tolist(), columns=[f'embedding{i+1}' for i in range(len(df['embedding'][0][0]))])
```

```python
embedding_df = pd.DataFrame(embeddings)
embedding_df.columns = [f'llama.embed{i+1}' for i in range(embeddings.shape[1])]
embedding_df = pd.concat([embedding_df, df[['score']]], axis=1)
```
