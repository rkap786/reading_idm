 Gunning Fog Index. It is a weighted average of the number of words per sentence, and the number of long words per word.

 The Daleâ€“Chall readability formula is a readability test that provides a numeric gauge of the comprehension difficulty that readers come upon when reading a text. It uses a list of 3000 words that groups of fourth-grade American students could reliably understand, considering any word not on that list to be difficult.

automated readability index (ARI):  approximate representation of the US grade level needed to comprehend the text. It uses number of letters, number of words, and number of sentences in text

Linsear Write - readability or grade level of text using easy or hard words, when easiness is defined by number of syllables in word

SMOG - count of polysyllabic words

SPACHE-  grade level based on average sentence length & complexity of text

```python

length= len(file)
print(length)
fk=[]
passage_len=[]
f= []
# gf=[]
# cl= []
# dale_chall=[]
# ari=[]
# linsear_write=[]
# smog=[]
# spache=[]

for i in range(length):
  text= file['Passage'][i]
  # print(i)
  if isinstance(text, str)==True:
    passage_len.append(len(text.split()))
    if len(text.split())<100:
      fk.append(np.nan)
      f.append(np.nan)
      # gf.append(np.nan)
      # cl.append(np.nan)
      # dale_chall.append(np.nan)
      # ari.append(np.nan)
      # linsear_write.append(np.nan)
      # smog.append(np.nan)
      # spache.append
      continue
    r = Readability(text)
    out = r.flesch_kincaid()
    fk.append(out.score)
    out= r.flesch()
    f.append(out.score)
    # out= r.gunning_fog()
    # gf.append(out.score)
    # out= r.coleman_liau()
    # cl.append(out.score)
    # out= r.dale_chall()
    # dale_chall.append(out.score)
    # out= r.ari()
    # ari.append(out.score)
    # out= r.linsear_write()
    # linsear_write.append(out.score)
    # out= r.smog()
    # smog.append(out.score)
    # out= r.spache()
    # spache.append(out.score)
```
