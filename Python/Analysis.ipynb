```python
f2 = load_dataset("domingue-lab/reading_idm",'data_qual_info')
f2 = f2['train'].to_pandas()
```

```python

f2= f2.drop([
    'Passage',
    'pdfname',
    'QuestionText',
    'PassNumUnq',
    'QuestionNo',
    'CorrectOptionNum',
    'QNoUnq',
    ], axis=1
)
f2= f2.loc[:,~f2.columns.str.startswith('option_')]
f2= f2.loc[:,~f2.columns.str.startswith('PerOption')]

mapping = {
    'Grade3': 3, 'Grade4': 4,
    'Grade5': 5, 'Grade6': 6,
    'Grade7': 7, 'Grade8': 8
}
f2.replace({'grade': mapping}, inplace=True)
mapping = {'NY': 0, 'Texas': 1}
f2.replace({'state': mapping}, inplace=True)
f2= f2.dropna()
f2_X= f2.drop('pVal', axis=1)
f2_Y= f2['pVal']
X_train, X_test, y_train, y_test = train_test_split(
    f2_X,f2_Y , random_state=123,test_size=0.20, shuffle=False
)
```

Split data into continous and categorical variables. Continous here include ordinal variables.

### Regression - prediction of pvalues

#### GPR

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e2))
gaussian_process = GaussianProcessRegressor(kernel=kernel, alpha=0, n_restarts_optimizer=10,)
gaussian_process.fit(X_train, y_train)
y_pred  = gaussian_process.predict(X_test)

print(root_mean_squared_error(y_test, y_pred))
np.corrcoef(y_pred, y_test)
```
